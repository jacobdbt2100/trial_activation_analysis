### Chi-square test of Independence

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Get activity columns from pivot table
activity_columns = [
    col for col in df_activity_pivot_flag.columns
    if col != 'ORGANIZATION_ID'
]

# Overall conversion rate (used internally only)
overall_conversion_rate = (
    df_organization['CONVERTED'].mean()
)

results = []

for activity in activity_columns:

    # Create contingency table
    contingency = pd.crosstab(
        df_organization_act_flag[activity],
        df_organization_act_flag['CONVERTED']
    )

    # Ensure full 2x2 structure
    contingency = contingency.reindex(
        index=[0, 1],
        columns=[False, True],
        fill_value=0
    )

    # Run Chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency)

    # Conversion rate after usage
    users_after_usage = contingency.loc[1].sum()

    if users_after_usage > 0:
        conversion_rate_after = (
            contingency.loc[1, True]
            / users_after_usage
        )
    else:
        conversion_rate_after = 0

    # Lift Ratio
    if overall_conversion_rate > 0:
        lift_ratio = (
            conversion_rate_after
            / overall_conversion_rate
        )
    else:
        lift_ratio = 0

    results.append({
        'ACTIVITY_NAME': activity,

        'CHI_SQUARE_STAT': round(chi2, 3),

        'P_VALUE': round(p_value, 5),

        'CONVERSION_RATE_AFTER_USAGE (%)':
            round(conversion_rate_after * 100, 1),

        'LIFT_RATIO':
            round(lift_ratio, 2)
    })

# Convert to DataFrame
chi_activity_results = pd.DataFrame(results)

# Sort by statistical significance
chi_activity_results = (
    chi_activity_results
    .sort_values('P_VALUE')
    .reset_index(drop=True)
)

chi_activity_results
```

### Mann–Whitney U test for Numeric Features

#### Outcome:
- All features had p-values > 0.05
- Shows **no statistically significant differences** between converted and non-converted users.

```python
import pandas as pd
from scipy.stats import mannwhitneyu

# Updated list of numeric columns (remove 'days_to_convert' due to nulls)
numeric_cols = [
    'trial_start_month',
    'total_events',
    'unique_activity_count',
    'active_days'
]

# Split into groups
group_true = df_organization[df_organization['CONVERTED'] == True]
group_false = df_organization[df_organization['CONVERTED'] == False]

results = []

# Loop through each feature
for col in numeric_cols:

    # Drop missing values
    g1 = group_true[col].dropna()
    g0 = group_false[col].dropna()

    # Skip empty columns
    if len(g1) == 0 or len(g0) == 0:
        continue

    # Mann–Whitney U test
    u_stat, u_p = mannwhitneyu(
        g1,
        g0,
        alternative='two-sided'
    )

    # Store results with mean and median
    results.append({
        'feature': col,

        'median_converted': round(g1.median(), 2),
        'median_not_converted': round(g0.median(), 2),

        'mean_converted': round(g1.mean(), 2),
        'mean_not_converted': round(g0.mean(), 2),

        'u_stat': round(u_stat, 2),
        'p_value': round(u_p, 5)
    })

# Convert to dataframe
results_df = pd.DataFrame(results)

# Sort by strongest statistical evidence
results_df = results_df.sort_values(
    by='p_value'
).reset_index(drop=True)

results_df
```

### Logistics Regression
- Derived fields only

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Target variable
y = df_organization['CONVERTED'].astype(int)

# Selected numeric features
numeric_features = [
    'total_events',
    'unique_activity_count',
    'active_days',
    'trial_start_month'
]

# Add activity features if needed later
activity_features = [
    # example:
    # 'CREATE_REPORT',
    # 'EXPORT_DATA'
]

selected_features = numeric_features + activity_features

# Create feature matrix
X = df_organization[selected_features].copy()

# Fill missing values
X = X.fillna(0)

# Convert nullable Int64 → float64 (critical fix)
X = X.astype('float64')

# Add intercept
X = sm.add_constant(X)

# Fit logistic regression
model = sm.Logit(y, X)
result = model.fit()

# Extract results
odds_ratios = np.exp(result.params)

logistic_results = pd.DataFrame({
    'Feature': result.params.index,
    'Coefficient': result.params.values,
    'Odds_Ratio': odds_ratios.values,
    'P_Value': result.pvalues.values
})

# Remove intercept
logistic_results = logistic_results[
    logistic_results['Feature'] != 'const'
]

# Sort by statistical importance
logistic_results = logistic_results.sort_values(
    by='P_Value'
).reset_index(drop=True)

logistic_results
```

### Logistics Regression
- All numeric fields (Derived fields & Activity usage count fields)

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Target
y = (
    df_organization_act_count['CONVERTED']
    .fillna(False)
    .astype(int)
)

# Columns to exclude
exclude_columns = [
    'ORGANIZATION_ID',
    'CONVERTED',
    'CONVERTED_AT',
    'TRIAL_START',
    'TRIAL_END',
    'days_to_convert'
]

# Select numeric features
numeric_features = (
    df_organization_act_count
    .drop(columns=exclude_columns, errors='ignore')
    .select_dtypes(include=['number'])
    .columns
)

X = df_organization_act_count[numeric_features].copy()

# Fill missing
X = X.fillna(0)

# Remove zero-variance columns
X = X.loc[:, X.nunique() > 1]

# Remove rare features
min_usage = 10

activity_usage = X.sum()

X = X.loc[:, activity_usage >= min_usage]

# Log-transform
X = np.log1p(X)

# Standardise
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# Fit Regularised Logistic Regression
model = LogisticRegression(
    penalty='l1',
    solver='liblinear',
    max_iter=1000
)

model.fit(X_scaled, y)

# Extract coefficients
coefficients = model.coef_[0]

odds_ratios = np.exp(coefficients)

logistic_results = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': coefficients,
    'Odds_Ratio': odds_ratios
})

# Sort by strongest effects
logistic_results = logistic_results.sort_values(
    by='Odds_Ratio',
    ascending=False
).reset_index(drop=True)

logistic_results
```

### Model AUC (CV = 5-fold stratified)

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
import matplotlib.pyplot as plt
import numpy as np

# Recreate model
model = LogisticRegression(
    penalty='l1',
    solver='liblinear',
    max_iter=1000
)

# 5-fold Stratified CV
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

# Compute ROC AUC
auc_scores = cross_val_score(
    model,
    X_scaled,
    y,
    cv=cv,
    scoring='roc_auc'
)

mean_auc = auc_scores.mean()

# Random baseline
random_baseline = 0.5

# Create vertical bar chart
plt.figure(figsize=(3,4))

models = ['Logistic Regression']
auc_values = [mean_auc]

plt.bar(
    models,
    auc_values,
    color='#2A9D8F'   # Attractive teal colour
)

# Random baseline line
plt.axhline(
    y=random_baseline,
    color='red',
    linestyle='--',
    label='Random Baseline (0.5)'
)

# Value label on bar
plt.text(
    0,
    mean_auc + 0.02,
    f'{mean_auc:.3f}',
    ha='center'
)

# Labels
plt.ylabel('ROC AUC')
plt.title('Model AUC (CV = 5-fold stratified)')

# Legend
plt.legend()

plt.ylim(0,1)

plt.show()
```

### Permutation Importance

```python
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.patches as mpatches

# Engagement features
engagement_features = [
    'total_events',
    'unique_activity_count',
    'active_days'
]

# IMPORTANT — Refit model
model.fit(X_scaled, y)

# Compute permutation importance
perm_importance = permutation_importance(
    model,
    X_scaled,
    y,
    n_repeats=20,
    random_state=42,
    scoring='roc_auc',
    n_jobs=-1
)

# Create results dataframe
perm_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance_Mean': perm_importance.importances_mean,
    'Importance_STD': perm_importance.importances_std
})

# ----------------------------
# DROP trial_start_month HERE
# ----------------------------
perm_df = perm_df[perm_df['Feature'] != 'trial_start_month']

# Sort
perm_df = perm_df.sort_values(
    by='Importance_Mean',
    ascending=False
)

# Top 20
top_n = 20

perm_top = perm_df.head(top_n).copy()

# Assign colours
perm_top['Color'] = perm_top['Feature'].apply(
    lambda x: '#2A9D8F'
    if x in engagement_features
    else '#B0B0B0'
)

# Reverse for horizontal plot
perm_top = perm_top.iloc[::-1]

# Plot
plt.figure(figsize=(7,5))

plt.barh(
    perm_top['Feature'],
    perm_top['Importance_Mean'],
    xerr=perm_top['Importance_STD'],
    color=perm_top['Color'],
    capsize=4
)

# Labels
plt.xlabel('Permutation Importance (ROC AUC decrease)')
plt.title('Top 20 Feature Importance (Permutation Importance)')

# Legend
eng_patch = mpatches.Patch(
    color='#2A9D8F',
    label='Engagement Features'
)

prod_patch = mpatches.Patch(
    color='#B0B0B0',
    label='Product Features'
)

plt.legend(handles=[eng_patch, prod_patch])

plt.tight_layout()

plt.show()
```

### Trial Goal 1:
- 5 or more **`Scheduling.Shift.Created`**

```python
# Define activity column
activity_col = 'Scheduling.Shift.Created'

# Define thresholds to test
thresholds = [1, 3, 5]

# Total number of organisations
total_orgs = df_organization_act_count.shape[0]

# Overall conversion rate
overall_conversion_rate = (
    df_organization_act_count['CONVERTED']
    .mean()
)

results = []

# Loop through thresholds
for t in thresholds:

    # Organisations meeting threshold
    used_df = df_organization_act_count[
        df_organization_act_count[activity_col] >= t
    ]

    # Number of organisations using
    num_orgs_used = used_df.shape[0]

    # Skip if no organisations meet threshold
    if num_orgs_used == 0:
        continue

    # Percent of organisations using
    pct_orgs_using = num_orgs_used / total_orgs

    # Number converted
    num_converted = used_df['CONVERTED'].sum()

    # Percent converted after usage
    pct_converted_after_usage = (
        num_converted / num_orgs_used
    )

    # Lift ratio
    lift_ratio = (
        pct_converted_after_usage /
        overall_conversion_rate
    )

    # Label formatting
    if t == 1:
        label = activity_col
    else:
        label = f'>= {t} {activity_col}'

    # Store results
    results.append({
        'ACTIVITY_LEVEL': label,

        'USERS_COUNT': num_orgs_used,

        'COMPLETION_RATE_%': round(
            pct_orgs_using * 100, 2
        ),

        'CONVERTED_AFTER_USAGE': num_converted,

        'CONVERSION_RATE_IF_COMPLETED_%': round(
            pct_converted_after_usage * 100, 2
        ),

        'LIFT_RATIO': round(lift_ratio, 2)
    })

# Convert to dataframe
activity_summary_df = pd.DataFrame(results)

activity_summary_df
```

```python
# Define activity column
activity_col = 'PunchClock.PunchedIn'

# Define thresholds
thresholds = [1, 3, 5]

# Total number of organisations
total_orgs = df_organization_act_count.shape[0]

# Overall conversion rate
overall_conversion_rate = (
    df_organization_act_count['CONVERTED']
    .mean()
)

results = []

# Loop through thresholds
for t in thresholds:

    # Organisations meeting threshold
    used_df = df_organization_act_count[
        df_organization_act_count[activity_col] >= t
    ]

    # Number of organisations using
    num_orgs_used = used_df.shape[0]

    # Skip if none meet threshold
    if num_orgs_used == 0:
        continue

    # Percent of organisations using
    pct_orgs_using = num_orgs_used / total_orgs

    # Number converted
    num_converted = used_df['CONVERTED'].sum()

    # Percent converted after usage
    pct_converted_after_usage = (
        num_converted / num_orgs_used
    )

    # Lift ratio
    lift_ratio = (
        pct_converted_after_usage /
        overall_conversion_rate
    )

    # Label formatting
    if t == 1:
        label = activity_col
    else:
        label = f'>= {t} {activity_col}'

    # Store results
    results.append({
        'ACTIVITY_LEVEL': label,

        'USERS_COUNT': num_orgs_used,

        'COMPLETION_RATE_%': round(
            pct_orgs_using * 100, 2
        ),

        'CONVERTED_AFTER_USAGE': num_converted,

        'CONVERSION_RATE_IF_COMPLETED_%': round(
            pct_converted_after_usage * 100, 2
        ),

        'LIFT_RATIO': round(lift_ratio, 2)
    })

# Convert to dataframe
activity_summary_df = pd.DataFrame(results)

activity_summary_df
```

### unique_activity_count_first_7_days

```python
# Ensure datetime formats
df_trial['TIMESTAMP'] = pd.to_datetime(df_trial['TIMESTAMP'])
df_trial['TRIAL_START'] = pd.to_datetime(df_trial['TRIAL_START'])

# -------- Create unique_activity_count_first_7_days -------- #

unique_activity_count_first_7_days = (
    df_trial[
        (df_trial['TIMESTAMP'] >= df_trial['TRIAL_START']) &
        (df_trial['TIMESTAMP'] <
         df_trial['TRIAL_START'] + pd.Timedelta(days=7))
    ]
    .groupby('ORGANIZATION_ID')['ACTIVITY_NAME']
    .nunique()
    .rename('unique_activity_count_first_7_days')
)

# -------- Temporary merge (not saved to original DF) -------- #

temp_df = (
    df_organization_act_count[
        ['ORGANIZATION_ID', 'CONVERTED']
    ]
    .merge(
        unique_activity_count_first_7_days,
        on='ORGANIZATION_ID',
        how='left'
    )
)

# Replace missing values with 0
temp_df['unique_activity_count_first_7_days'] = (
    temp_df['unique_activity_count_first_7_days']
    .fillna(0)
)

# -------- Threshold Analysis -------- #

# Added >= 2 here
thresholds = [1, 2, 3, 5, 10]

total_orgs = temp_df.shape[0]

overall_conversion_rate = (
    temp_df['CONVERTED']
    .mean()
)

results = []

for t in thresholds:

    used_df = temp_df[
        temp_df['unique_activity_count_first_7_days'] >= t
    ]

    num_orgs_used = used_df.shape[0]

    if num_orgs_used == 0:
        continue

    pct_orgs_using = num_orgs_used / total_orgs

    num_converted = used_df['CONVERTED'].sum()

    pct_converted_after_usage = (
        num_converted / num_orgs_used
    )

    lift_ratio = (
        pct_converted_after_usage /
        overall_conversion_rate
    )

    if t == 1:
        label = 'unique_activity_count_first_7_days'
    else:
        label = f'>= {t} unique_activity_count_first_7_days'

    results.append({
        'ACTIVITY_LEVEL': label,

        'USERS_COUNT': num_orgs_used,

        'COMPLETION_RATE_%': round(
            pct_orgs_using * 100, 2
        ),

        'CONVERTED_AFTER_USAGE': num_converted,

        'CONVERSION_RATE_IF_COMPLETED_%': round(
            pct_converted_after_usage * 100, 2
        ),

        'LIFT_RATIO': round(lift_ratio, 2)
    })

unique_activity_7d_summary_df = pd.DataFrame(results)

unique_activity_7d_summary_df
```
